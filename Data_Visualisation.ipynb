# **Grouping the Data set**

```python
data_bar = data[['JobRole','MonthlyIncome']]
data_bar = data_bar.groupby(['JobRole'], as_index=False).mean()
data_bar
```

# **Bar Charts using Matplotlib / Seaborn / Plotly**

```python
data_bar.plot.barh(x='JobRole', y='MonthlyIncome')

plt.xlabel('Average Monthly Income') # add to x-label to the plot
plt.ylabel('Job Role') # add y-label to the plot
plt.title('Average Monthly Income on a Job Role') # add title to the plot
plt.legend(['Average Monthly Income'])

plt.show()
```

```python
sns.barplot(x='MonthlyIncome', y='JobRole', data=data_bar).set_title("Average Monthly Income on a Job Role")
plt.xlabel("Average Monthly Income")
plt.ylabel("Job Role")
```

```python

fig = px.bar(data_bar, x='JobRole', y='MonthlyIncome',  color='MonthlyIncome', title='Average Monthly Income on a Job Role' )
fig.show()
```

# **Advance Charts**

```python
data_waf = data[data['JobRole']=='Manager']
data_waf = data_waf[['EducationField','MonthlyIncome']]
data_waf = data_waf.groupby(['EducationField'], as_index=False).count()
data_waf.rename(columns={'MonthlyIncome': 'Number of Managers'}, inplace=True)
data_waf
```

```python
%pip install pywaffle
from pywaffle import Waffle
```

```python
df = pd.DataFrame(data_waf)

# compute percentage values
percentages = (data_waf['Number of Managers'] / data_waf['Number of Managers'].sum()) * 100

# create labels with percentage values
labels = [f"{label} ({percentage:.1f}%)" for label, percentage in zip(data_waf['EducationField'], percentages)]

# create a waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=6,
    columns=17,
    values=df['Number of Managers'],
    labels=labels,
    figsize=(8, 6),
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    colors=['#0072C6', '#CE1126', '#FDB913', '#00BFFF', '#000080','#00FF00']
)
plt.title('Field of Education of Managers')
plt.show()
```

```python
# Create the regression plot
sns.regplot(x='Age', y='MonthlyIncome', scatter_kws={"color": "black"}, line_kws={"color": "red"}, data=data)

# Add a title to the plot
plt.title('Regression Plot of Monthly Income vs Age')
# Set x-axis label
plt.xlabel('Age')
# Set y-axis label
plt.ylabel('Monthly')
# Display the plot
plt.show()
```
