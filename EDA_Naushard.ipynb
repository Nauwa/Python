**Importing Packages**

```python
import pandas as pd
import numpy as np
```

**Creating the Data Frame for CSV input**

```python
data = pd.read_csv('imdb_movie_data.csv')
data
```

**Data Preprocessing**

```python
data.replace("'-", np.nan, inplace = True)
```

```python
data.dropna(inplace=True)
```

```python
data['year'] = data['year'].str.replace(r'[()]',"")
```

```python
data['certificate'] = data['certificate'].str.replace(r'[()]',"")
data
```

```python
data['gross_earning'] = data['gross_earning'].str.replace(r'[M]',"")
data
```

```python
data['gross_earning'] = data['gross_earning'].str.replace(r'[$]',"")
data
```

```python
data['gross_earning'] = data['gross_earning'].str.replace(r'[#]',"")
data
```

```python
data['certificate'] = data['certificate'].str.replace(r'[+]',"")
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data.gross_earning = data.gross_earning.astype(float)
```

```python
data.dtypes
```

```python
data['time_minute'] = data['time_minute'].str.replace(r'[ MIN]',"")
data
```

```python
data['time_minute'] = data['time_minute'].str.replace(r'[min]',"")
data
```

```python
data.dtypes
```

```python
data.time_minute = data.time_minute.astype(int)
```

```python
data.dtypes
```

```python
data.reset_index(inplace=True)
```

```python
data.year = data.year.astype(int)
bins = np.linspace(min(data['year']), max(data['year']), 8)
bins
```

```python
group_names = ['1', '2', '3','4','5','6','7']
```

```python
data['year-binned'] = pd.cut(data['year'], bins, labels=group_names , include_lowest=True )
data[['year','year-binned']].head(5)
```

```python
data["year-binned"].value_counts()
```

```python
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, data['year-binned'].value_counts())

plt.pyplot.xlabel("Year Bin")
plt.pyplot.ylabel("Movie Count")
plt.pyplot.title("Movie count in Year Bin")
```

```python
# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(data["year"], bins = 6)

# set x/y labels and plot title
plt.pyplot.xlabel("Year Bin")
plt.pyplot.ylabel("count")
plt.pyplot.title("Movie count in Year Bin")
```
